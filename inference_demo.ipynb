```python
from mmdet.apis import init_detector, inference_detector, show_result_pyplot,get_result
import mmcv
```

```python
config_file = './myconfigs/cascade_rcnn_dconv_c3-c5_x101_32x4d_fpn_snms_1x_fintune.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
checkpoint_file = './work_dirs/cascade_rcnn_dconv_c3-c5_x101_32x4d_fpn_snms_1x_fintune/epoch_2.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single image
img = './data/safehat/JPEGImages/000002.jpg'
result = inference_detector(model,img)
img = get_result(img,result,class_names=model.CLASSES,auto_thickness=True,color_dist={'hat':'red','person':'green'})
```

```python
# show the results
from matplotlib import pyplot as plt
import cv2
```
